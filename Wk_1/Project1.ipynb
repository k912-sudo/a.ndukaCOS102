```python
# Simple interest

principal = float(input("What is the principal"))
print('The principal is',(principal))

rate = float(input("What is the rate"))
r = rate / 100
print('The rate is',(r))
      
time = float(input("What is the amount of time"))
print('The time is',(time))

interest = principal * r * time
print("The simple interest is", (interest))
```
