```python
# Compound interest

p = float(input("What is the principal"))
print("The principal is ",(p))

r = float(input("What is the r"))
print("The rate is ",(r))

t = float(input("What is the time"))
print("The time is ",(t))

n = float(input("What is the frequency"))
print("The frequency is ",(n))

A = float(p * (1.0 + r / n) ** (n*t))
print("The amount is",(A))

CI = A - p
print("The compound interest is",(CI))
```
